### Image Processing Techniques:
Image Filtering

Image filtering is a fundamental technique in image processing, used to enhance or modify images. Here are some common filtering techniques:

Blurring: Blurring, or smoothing, is used to reduce noise or detail in an image. You can use different types of blurs, such as Gaussian blur, median blur, or average blur. For instance, if you’re developing an application to anonymize faces, you might use a Gaussian blur to obscure details.


```python
cat_image=cv2.imread('cat.jpg')

# Convert BGR to RGB for display
cat_image = cv2.cvtColor(cat_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(cat_image)
plt.axis('off')
plt.title("cat Image")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(cat_image, (7, 7), 5)
# Show the result
plt.imshow(blurred_image)
plt.axis('off')
plt.title("cat Image")
plt.show()
```

Sharpening: Sharpening enhances the edges of objects in an image. It’s useful in scenarios like medical imaging, where enhancing the edges can help in better diagnosis

```python
chest_image=cv2.imread('hands.jpg')

# Convert BGR to RGB for display
# chest_image = cv2.cvtColor(chest_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(chest_image)
plt.axis('off')
plt.title("xray Image")
plt.show()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
sharpened_image = cv2.filter2D(chest_image, -1, kernel)
```

```python
# Convert BGR to RGB for display
sharpened_rgb = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(sharpened_rgb)
plt.axis('off')
plt.title("Sharpened Image")
plt.show()
```

Thresholding

Thresholding is a technique used to create binary images from grayscale images. It’s widely used in document scanning and object segmentation.

Binary Thresholding: This is the simplest form, where each pixel is set to 255 if its value is above a certain threshold, and 0 otherwise.

```python
ret, thresh_binary = cv2.threshold(chest_image, 127, 255, cv2.THRESH_BINARY)
display_image(thresh_binary)
```

Morphological Operations

Morphological operations process images based on shapes. They are particularly useful in removing noise and enhancing object structures.

Erosion and Dilation: Erosion removes pixels at the boundaries of objects, while dilation adds pixels to the boundaries. They’re often used together for noise reduction.


```python
text_image=cv2.imread('text.png')

# Convert BGR to RGB for display
text_image = cv2.cvtColor(text_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(text_image)
plt.axis('off')
plt.title("text Image")
plt.show()
```

```python
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(text_image, kernel, iterations=1)
dilation = cv2.dilate(text_image, kernel, iterations=1)
```

```python
# Show the result
plt.imshow(erosion)
plt.axis('off')
plt.title("erosion Image")
plt.show()
```

```python
# Show the result
plt.imshow(dilation)
plt.axis('off')
plt.title("dilation Image")
plt.show()
```

Opening and Closing: Opening is erosion followed by dilation, useful for removing small objects. Closing is dilation followed by erosion, helpful in closing small holes within objects.



```python
noise_image=cv2.imread('noise.png')

# Convert BGR to RGB for display
noise_image = cv2.cvtColor(noise_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(noise_image)
plt.axis('off')
plt.title("noise Image")
plt.show()
```

```python
opening = cv2.morphologyEx(noise_image, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(noise_image, cv2.MORPH_CLOSE, kernel)
```

```python
# Show the result
plt.imshow(opening)
plt.axis('off')
plt.title("opening Image")
plt.show()
```

```python
# Show the result
plt.imshow(closing)
plt.axis('off')
plt.title("closing Image")
plt.show()
```

```python
noise_image=cv2.imread('13.png')

# Convert BGR to RGB for display
noise_image = cv2.cvtColor(noise_image, cv2.COLOR_BGR2RGB)

# Show the result
plt.imshow(noise_image)
plt.axis('off')
plt.title("noise Image")
plt.show()
```

```python
def noise_removal(image):
    import numpy as np
    kernel = np.ones((1, 1), np.uint8)
    # This line applies dilation to the input image using the created kernel.
    # Dilation is an operation that expands the white regions of the image.
    # It can be used to fill in small gaps and enlarge regions.
    image = cv2.dilate(image, kernel, iterations=1)
    # This line applies erosion to the image using the kernel.
    # Erosion is an operation that shrinks the white regions of the image.
    # It's used to remove small islands of noise.
    image = cv2.erode(image, kernel, iterations=1)
    # these are †he main methods that remove noise 
    image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    # This line applies a median blur to the image with a kernel size of 3x3.
    # Median blur is a type of smoothing filter that replaces each pixel's value
    # with the median value of its neighboring pixels.
    # It's effective in reducing noise while preserving edges.
    image = cv2.medianBlur(image, 5)
    return (image)
```

```python
denoised_image = noise_removal(noise_image)
```

```python
# Show the result
plt.imshow(denoised_image)
plt.axis('off')
plt.title("denoised Image")
plt.show()
```

Edge Detection

The Canny edge detector is a popular algorithm for detecting edges in images. It’s used in applications like handwriting recognition and fingerprint analysis.



```python
from PIL import Image
import numpy as np

# Use PIL to open the GIF
gif_image = Image.open("shakey.jpg")

# Convert to RGB and then to NumPy array
gif_rgb = gif_image.convert("RGB")
edges_image = np.array(gif_rgb)

# Display using matplotlib
plt.imshow(edges_image)
plt.axis('off')
plt.title("GIF Image")
plt.show()
```

```python
edges = cv2.Canny(edges_image, 100, 200)
```
